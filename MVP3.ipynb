### 2. Modelagem

O nome da tabela será *base_cosumo* e contará com as seguintes colunas:

*   ID (int): Identificador único para cada um indivíduo
*   year_birth (date): Data de nascimento

*   marital_status (string): Estado Civil
*   income (int): Renda familiar
*   kidhome(int): Qtd de filhos crianças
*   teenhome (int): Qtd de filhos adolescentes
